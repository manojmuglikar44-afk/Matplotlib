```python
import matplotlib.pyplot as plt
import numpy as np 

Apps = np.array(["Instagram","YouTube","Spotify","whatsApp","Others"])
hours = np.array([3 ,5 ,1 ,2 ,6])

plt.title("Daily Phone Usage")
plt.xlabel("Apps")
plt.ylabel("hours")

plt.bar(Apps,hours, color = "cyan")
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np 

Apps = np.array(["Instagram","YouTube","Spotify","whatsApp","Others"])
hours = np.array([3 ,5 ,1 ,2 ,6])

plt.pie(hours, labels = Apps,
               autopct = "%.1f%%")
plt.title("Daily Phone Usage", fontsize = 20,
                            family = "Times New Roman")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 

Apps = np.array(["Instagram","YouTube","Spotify","whatsApp","Others"])
hours = np.array([3 ,5 ,1 ,2 ,6])

plt.pie(hours, labels = Apps,
               autopct = "%.1f%%",
               explode = [0, 0, 0.1, 0, 0.1],
               shadow = True,
               startangle = 180)
plt.title("Daily Phone Usage", fontsize = 20,
                            family = "Times New Roman")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 

x1 = [0,1,2,2,3,4,5,6,7,7,8,8]
y1 = [50,68,65,60,62,78,83,78,92,95,97,45]
y2 = [52,70,70,50,65,75,80,80,90,95,100,45]

plt.scatter(x1,y1, color = "cyan",
                   alpha = 0.5,
                   s = 50,
                   label = "Class A")
plt.scatter(x1,y2, color = "red",
                   alpha = 0.5,
                   s = 50,
                   label = "Class B")

plt.title("College",fontsize = 20)
plt.xlabel("Hours Studied")
plt.ylabel("Grade")

plt.legend()
plt.show()
```
