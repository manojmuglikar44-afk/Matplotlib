```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5])

figures, axes = plt.subplots(2,2)

axes[0,0].plot(x,x**2, color = "red")

axes[0,1].plot(x,x**3, color = "green")

axes[1,0].plot(x,x**4, color = "green")

axes[1,1].plot(x,x**5, color = "green")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5])

figures, axes = plt.subplots(2,2)

axes[0,0].plot(x,x**2, color = "red")
axes[0,0].set_title("X**2")

axes[0,1].plot(x,x**3, color = "blue")
axes[0,1].set_title("X**3")

axes[1,0].plot(x,x**4, color = "green")
axes[1,0].set_title("X**4")

axes[1,1].plot(x,x**5, color = "purple")
axes[1,1].set_title("X**5")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5])

figures, axes = plt.subplots(2,2)

axes[0,0].plot(x,x**2, color = "red")
axes[0,0].set_title("X**2")

axes[0,1].plot(x,x**3, color = "blue")
axes[0,1].set_title("X**3")

axes[1,0].plot(x,x**4, color = "green")
axes[1,0].set_title("X**4")

axes[1,1].plot(x,x**5, color = "purple")
axes[1,1].set_title("X**5")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5])

figures, axes = plt.subplots(2,2)

axes[0,0].plot(x,x**2, color = "red")
axes[0,0].set_title("X**2")

axes[0,1].bar(x,x**3, color = "blue")
axes[0,1].set_title("X**3")

axes[1,0].bar(x,x**4, color = "green")
axes[1,0].set_title("X**4")

axes[1,1].plot(x,x**5, color = "purple")
axes[1,1].set_title("X**5")

plt.tight_layout()
plt.show()
```
